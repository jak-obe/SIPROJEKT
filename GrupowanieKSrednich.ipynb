```python
import pandas as pd
```

```python
data = pd.read_excel("datasets/Raisin_Dataset.xlsx")
X = data.drop('Class', axis=1)  
y = data['Class'] 
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X) 
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))  # Test clusters from 2 to 10
visualizer.fit(X)    
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=2,  # Replace 3 with your chosen number of clusters 
                init='k-means++', # Smart initialization
                random_state=42)  # Set random_state for reproducibility
```

```python
kmeans.fit(scaled_data)
```

```python
labels = kmeans.labels_
```

```python
import matplotlib.pyplot as plt

plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=labels, cmap='viridis')
plt.title('K-Means Clustering Results')
plt.colorbar()  # Add a colorbar to show the color-label mapping
plt.show()
```
