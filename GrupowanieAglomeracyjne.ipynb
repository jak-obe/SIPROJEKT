```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("datasets/Raisin_Dataset.xlsx")
X = data.drop('Class', axis=1)
y = data['Class']
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)
```

```python
clustering2 = AgglomerativeClustering(n_clusters=2)

plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clustering2.fit_predict(scaled_data), cmap='viridis')
plt.title('Metoda Hierarchiczna Aglomeracyjna')
plt.colorbar()  # Add a colorbar to show the color-label mapping
plt.show()
```
